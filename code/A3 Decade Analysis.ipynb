```python
def global_measure(word:str, modelA, modelB):
  try:
    A_wv = modelA.wv[word].reshape(1,-1)
    B_wv = modelB.wv[word].reshape(1,-1)

    output = cosine_similarity(A_wv, B_wv)
    return output[0][0]
  except:
    return None

def local_neighborhood_measure(word:str, modelA, modelB):
  # get K-nearest-neighbors where K=50
  A_similarity_pairs = modelA.wv.most_similar(word, topn=50)
  B_similarity_pairs = modelB.wv.most_similar(word, topn=50)

  # get the similar words
  A_similarity_words = [words for words, similarity_score in A_similarity_pairs]
  B_similarity_words = [words for words, similarity_score in B_similarity_pairs]

  # get the union of the most similar words from both the old and modern set
  union_words = A_similarity_words + B_similarity_words
  # deduplicate words by converting the list into a set and back into a list
  union_words = list(set(union_words))

  # for each model get the similarity score of the target words to each word in the list of union words
  # this list of cosine similarity scores can be treated as a second order embedding
  old_second_order_embedding = np.array([modelA.wv.similarity(word, w) for w in union_words]).reshape(1,-1)
  modern_second_order_embedding = np.array([modelB.wv.similarity(word, w) for w in union_words]).reshape(1,-1)

  # compare the cosine similarity between the two
  output = cosine_similarity(old_second_order_embedding, modern_second_order_embedding)
  return output[0][0]
```

```python
global_measure_scores = {}

for word in target_words:
  global_measure_scores[word] = global_measure(word, model_1900s, model_1910s)

for word, score in global_measure_scores.items():
  print(f'{word}: Global Measure = {score}')
```
